## Data cleaning

### Subtask:
Clean the dataset by handling the zero values in the 'Volume' column and addressing any potential outliers in the numerical features.


**Reasoning**:
Investigate and handle zero volumes in the 'Volume' column and address outliers in numerical features.  This will involve visualizations, imputation/removal of zero volumes, outlier detection, and data validation.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Investigate Zero Volumes
zero_volume_df = df[df['Volume'] == 0]
print(f"Number of rows with zero volume: {len(zero_volume_df)}")
display(zero_volume_df.head())

plt.figure(figsize=(12, 6))
for index in df['Index'].unique():
    index_df = df[df['Index'] == index]
    plt.plot(index_df['Date'], index_df['Volume'], label=index)
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume Over Time for Different Indexes')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Handle Zero Volumes (Imputation using median volume for each index)
for index in df['Index'].unique():
    median_volume = df[df['Index'] == index]['Volume'].median()
    df.loc[(df['Index'] == index) & (df['Volume'] == 0), 'Volume'] = median_volume

# Outlier Detection and Handling (Winsorizing)
numerical_cols = ['Open', 'High', 'Low', 'Close']
for col in numerical_cols:
    upper_limit = df[col].quantile(0.99)
    lower_limit = df[col].quantile(0.01)
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

# Data Validation (Recalculate descriptive statistics and visualize distributions)
print("\nDescriptive Statistics after cleaning:")
display(df.describe())

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols + ['Volume']):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col} after cleaning')
plt.tight_layout()
plt.show()


# Save cleaned data
df.to_csv("nse_indexes_cleaned.csv", index=False)
```

## Feature engineering

### Subtask:
Engineer new features from the cleaned dataset.


**Reasoning**:
Load the cleaned dataset and engineer new features, then save the engineered dataset.



```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the cleaned dataset
try:
    df = pd.read_csv('nse_indexes_cleaned.csv')
except FileNotFoundError:
    print("Error: 'nse_indexes_cleaned.csv' not found.")
    df = None
    # Handle the error appropriately, e.g., exit or use a default DataFrame

if df is not None:
    # Convert 'Date' column to datetime objects if it's not already
    if not pd.api.types.is_datetime64_any_dtype(df['Date']):
        df['Date'] = pd.to_datetime(df['Date'])

    # Feature Engineering
    df = df.sort_values(by='Date')  # Ensure data is sorted by date
    df['Close_Lag1'] = df['Close'].shift(1)
    df['Close_Lag2'] = df['Close'].shift(2)
    df['Close_Lag5'] = df['Close'].shift(5)
    df['Close_Lag10'] = df['Close'].shift(10)

    df['Close_MA7'] = df['Close'].rolling(window=7).mean()
    df['Close_MA14'] = df['Close'].rolling(window=14).mean()
    df['Close_MA30'] = df['Close'].rolling(window=30).mean()

    df['Daily_Return'] = df['Close'].pct_change()
    df['High_Low_Spread'] = df['High'] - df['Low']

    # Drop rows with NaN values created by lagging
    df.dropna(inplace=True)

    # Scale numerical features
    scaler = MinMaxScaler()
    numerical_features = ['Open', 'High', 'Low', 'Close', 'Volume', 'Close_Lag1', 'Close_Lag2', 'Close_Lag5', 'Close_Lag10', 'Close_MA7', 'Close_MA14', 'Close_MA30', 'Daily_Return', 'High_Low_Spread']
    df[numerical_features] = scaler.fit_transform(df[numerical_features])


    # Save the engineered dataset
    df.to_csv('nse_indexes_engineered.csv', index=False)
    print("Successfully saved the engineered dataset to nse_indexes_engineered.csv")
```

## Data splitting

### Subtask:
Split the engineered dataset into training, validation, and testing sets.


**Reasoning**:
Split the data into training, validation, and testing sets using a temporal split.



```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Define features (X) and target variable (y)
X = df.drop(columns=['Close', 'Date', 'Index', 'Currency'])
y = df['Close']

# Calculate split indices
train_size = int(len(X) * 0.8)
val_size = int(len(X) * 0.1)
test_size = len(X) - train_size - val_size

# Split the data
X_train = X[:train_size]
y_train = y[:train_size]
X_val = X[train_size:train_size + val_size]
y_val = y[train_size:train_size + val_size]
X_test = X[train_size + val_size:]
y_test = y[train_size + val_size:]

# Ensure all dataframes have same index
X_train = X_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
X_val = X_val.reset_index(drop=True)
y_val = y_val.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

# Check shapes of splits
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

## Model optimization

### Subtask:
Optimize the hyperparameters of the Linear Regression, Ridge Regression, and KNN models using the validation set (X_val, y_val).  For Q-learning, since it's a placeholder, skip the optimization step.


**Reasoning**:
Optimize the hyperparameters of Ridge Regression and KNN models using the validation set and refit them on the combined training and validation sets.  Linear Regression does not have hyperparameters to tune. Skip hyperparameter optimization for the Q-learning model as instructed.



```python
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Linear Regression (no hyperparameter tuning)
y_pred_lr = linear_regression_model.predict(X_val)
rmse_lr = np.sqrt(mean_squared_error(y_val, y_pred_lr))
print(f"Linear Regression RMSE on validation set: {rmse_lr}")


# Ridge Regression
alpha_values = np.linspace(0.01, 10, 50)  # Try 50 alpha values between 0.01 and 10
best_alpha = None
best_rmse = float('inf')
for alpha in alpha_values:
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X_train, y_train)
    y_pred_ridge = ridge_model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred_ridge))
    if rmse < best_rmse:
        best_rmse = rmse
        best_alpha = alpha

print(f"Best alpha for Ridge Regression: {best_alpha}")
print(f"Ridge Regression RMSE on validation set: {best_rmse}")

# Refit Ridge Regression with best alpha
X_train_combined = pd.concat([X_train, X_val])
y_train_combined = pd.concat([y_train, y_val])
ridge_regression_model = Ridge(alpha=best_alpha)
ridge_regression_model.fit(X_train_combined, y_train_combined)


# KNN
n_neighbors_values = np.arange(3, 16)
best_k = None
best_rmse_knn = float('inf')
for k in n_neighbors_values:
  knn_model = KNeighborsRegressor(n_neighbors=k)
  knn_model.fit(X_train, y_train)
  y_pred_knn = knn_model.predict(X_val)
  rmse = np.sqrt(mean_squared_error(y_val, y_pred_knn))
  if rmse < best_rmse_knn:
      best_rmse_knn = rmse
      best_k = k

print(f"Best k for KNN: {best_k}")
print(f"KNN RMSE on validation set: {best_rmse_knn}")

# Refit KNN with the best k
knn_model = KNeighborsRegressor(n_neighbors=best_k)
knn_model.fit(X_train_combined, y_train_combined)
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained Linear Regression, Ridge Regression, KNN, and Q-Learning models on the test set (X_test, y_test) using RMSE and MAE.


**Reasoning**:
Evaluate the performance of the trained models using RMSE and MAE on the test set.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Predict on the test set
y_pred_lr = linear_regression_model.predict(X_test)
y_pred_ridge = ridge_regression_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
y_pred_ql = q_learning_model.predict(X_test)


# Calculate RMSE and MAE
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
mae_lr = mean_absolute_error(y_test, y_pred_lr)

rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)

rmse_knn = np.sqrt(mean_squared_error(y_test, y_pred_knn))
mae_knn = mean_absolute_error(y_test, y_pred_knn)

rmse_ql = np.sqrt(mean_squared_error(y_test, y_pred_ql))
mae_ql = mean_absolute_error(y_test, y_pred_ql)


# Store results
model_performance = {
    'Linear Regression': {'RMSE': rmse_lr, 'MAE': mae_lr},
    'Ridge Regression': {'RMSE': rmse_ridge, 'MAE': mae_ridge},
    'KNN': {'RMSE': rmse_knn, 'MAE': mae_knn},
    'Q-Learning': {'RMSE': rmse_ql, 'MAE': mae_ql},
}

# Print results
print("Model Performance:")
for model_name, metrics in model_performance.items():
    print(f"{model_name}:")
    print(f"  RMSE: {metrics['RMSE']:.4f}")
    print(f"  MAE: {metrics['MAE']:.4f}")
```
